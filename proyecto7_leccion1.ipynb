# Proyecto: Segmentador Inteligente de Clientes Minoristas
## Lección 1: Fundamentos del Aprendizaje No Supervisado

### A. Tareas principales del aprendizaje no supervisado

En el aprendizaje no supervisado, el objetivo es descubrir estructuras, patrones o relaciones ocultas en los datos sin contar con una variable objetivo (etiqueta) que guíe el proceso. Para nuestro proyecto en **Retail Insights S.A.**, estas son las tareas fundamentales que aplicaremos:

1. **Agrupamiento (Clustering):**  Es la tarea de organizar los datos en grupos con características similares. Nos permitirá identificar segmentos de clientes (por ejemplo: "compradores frecuentes" o "clientes de bajo gasto") basándonos únicamente en sus comportamientos.
2. **Reducción de Dimensionalidad:**  Esta técnica simplifica el conjunto de datos al reducir el número de variables (columnas) sin perder la información más importante. Es vital para limpiar el "ruido" y poder visualizar a los clientes en gráficos claros.
3. **Detección de Anomalías:** Identifica puntos de datos que se desvían significativamente del comportamiento normal. Es útil para encontrar errores en la carga de datos o comportamientos de compra inusuales que podrían sesgar el análisis.
4. **Reglas de Asociación:** Descubre relaciones entre variables, como saber qué productos suelen comprarse juntos. Esto ayuda a diseñar promociones o mejorar la disposición de las tiendas.

### B. Clasificación de técnicas por tipo

Para el desarrollo del proyecto en **Retail Insights S.A.**, es necesario organizar las herramientas de aprendizaje no supervisado según su objetivo analítico. Estas se dividen principalmente en tres categorías:

#### 1. Clusterización (Agrupamiento)
Estas técnicas buscan dividir el conjunto de datos en grupos (clústeres) donde los elementos internos sean muy similares entre sí y muy diferentes a los de otros grupos.
* **K-Means:** Divide los datos en un número "K" de grupos basándose en la distancia a centros geométricos (centroides).
* **DBSCAN:** Agrupa puntos basándose en la densidad; es ideal para encontrar grupos de formas irregulares y detectar valores atípicos.
* **Agrupamiento Jerárquico:** Crea una jerarquía de grupos que se puede visualizar en un diagrama de árbol (dendrograma).



#### 2. Reducción de Dimensionalidad
Se utilizan para simplificar datasets con muchas variables, resumiendo la información en unas pocas dimensiones principales sin perder la esencia de los datos.
* **PCA (Análisis de Componentes Principales):** Técnica lineal que proyecta los datos en nuevas dimensiones que capturan la mayor variabilidad posible.
* **t-SNE:** Técnica no lineal diseñada **exclusivamente para la visualización** de datos complejos en 2D o 3D, manteniendo las relaciones de cercanía entre puntos. ⚠️ A diferencia de PCA, t-SNE **no debe usarse como preprocesamiento para modelos de clustering**, ya que no preserva distancias globales y sus coordenadas de salida no son deterministas. Su rol en el proyecto es comunicar visualmente los segmentos detectados, no generarlos.



#### 3. Asociación
Buscan descubrir reglas y relaciones ocultas entre variables en grandes volúmenes de datos transaccionales.
* **Algoritmo Apriori:** Identifica conjuntos de elementos que aparecen juntos con frecuencia (ej. "clientes que compran café también compran azúcar").
* **Eclat:** Una alternativa más rápida a Apriori que utiliza intersecciones de conjuntos para encontrar patrones frecuentes.

### C. Analizar casos reales de aplicación

El aprendizaje no supervisado no es solo teoría; es la base de muchas funciones que usamos a diario. Aquí te presento cómo se aplican las técnicas que usaremos en escenarios del mundo real:

#### 1. Segmentación de Clientes (Clustering)
Es el caso central de nuestro proyecto para **Retail Insights**. 
* **Aplicación:** Una cadena de supermercados utiliza K-Means para agrupar a sus clientes no solo por lo que gastan, sino por la hora en la que compran y el tipo de productos (frescos vs. procesados).
* **Resultado:** Permite enviar cupones de descuento personalizados. Por ejemplo, si el algoritmo identifica un grupo de "compradores de desayuno saludable", se les envían ofertas de avena y frutas.

#### 2. Compresión de Imágenes y Reconocimiento Facial (Reducción de Dimensionalidad)
* **Aplicación:** Los sistemas de seguridad que utilizan reconocimiento facial manejan imágenes con miles de píxeles (dimensiones). Usar todos esos datos sería muy lento.
* **Resultado:** Mediante **PCA**, el sistema "comprime" la imagen quedándose solo con los rasgos más importantes (distancia entre ojos, forma de la mandíbula). Esto permite que el algoritmo compare caras en milisegundos sin perder precisión.


#### 3. Análisis de Canasta de Compras (Asociación)
* **Aplicación:** Amazon o Netflix utilizan estas reglas para sus motores de recomendación.
* **Resultado:** Si el sistema detecta que el 80% de las personas que compran una "Cámara Reflex" también compran una "Memoria SD" y un "Trípode", creará automáticamente una sección de "Comprados juntos habitualmente". Esto aumenta el ticket promedio de venta sin intervención humana.


#### 4. Bioinformática y Genética (Clustering y Reducción)
* **Aplicación:** En la medicina de precisión, se analizan miles de genes al mismo tiempo para entender enfermedades complejas.
* **Resultado:** Se utiliza **t-SNE** para visualizar grupos de pacientes con expresiones genéticas similares, lo que ayuda a los científicos a identificar qué grupos responden mejor a un tratamiento específico y cuáles no.
