# Informe Final: Segmentación Inteligente de Clientes
**Cliente:** Retail Insights S.A.
**Analista:** Sebastian Huichal

## 1. Introducción y Objetivo del Proyecto
Retail Insights S.A. identificó la necesidad de abandonar las estrategias de marketing genérico masivo para uno de sus clientes principales, una cadena con presencia en tiendas físicas y e-commerce. Hasta la fecha, las campañas se basaban en información básica, perdiendo la oportunidad de capitalizar los hábitos reales de consumo.

El objetivo de este proyecto fue desarrollar un modelo de segmentación basado en algoritmos de aprendizaje no supervisado. Al analizar una base de 2000 clientes con 11 variables de comportamiento, demográficas y de interacción digital, el sistema logró descubrir patrones ocultos para crear perfiles accionables, permitiendo la personalización de futuras estrategias comerciales.

## 2. Metodología y Tratamiento de Datos
Para garantizar que el análisis fuera preciso y escalable, se aplicó un flujo de trabajo riguroso:

* **Limpieza y Preparación:** Se estructuró un conjunto de datos multidimensional que incluía ingresos, frecuencia de visitas, gasto por categoría (electrónica, hogar, ropa, alimentos, entre otros) y métricas de uso de la aplicación móvil.
* **Reducción de Dimensionalidad (PCA):** Trabajar con 11 variables simultáneas genera ruido computacional. Se aplicó el Análisis de Componentes Principales (PCA), conservando únicamente los componentes necesarios para explicar más del 80% de la varianza de los datos. 
* **Visualización (t-SNE):** Para que los resultados fuesen interpretables visualmente, se utilizó la técnica t-SNE, proyectando la complejidad multidimensional en gráficos 2D claros.

## 3. Selección y Validación del Modelo Maestro
Se pusieron a prueba tres enfoques de clusterización:

1.  **Agrupamiento Jerárquico:** Descartado para el modelo final por su alto consumo de memoria, lo que limitaba su escalabilidad futura.
2.  **DBSCAN:** Descartado porque su mecánica de aislar clientes atípicos como ruido terminó fragmentando grupos que comercialmente tenían sentido y afectó negativamente el puntaje de cohesión general al evaluarlos con el Coeficiente de Silueta.
3.  **K-Means (Modelo Seleccionado):** Elegido como el algoritmo definitivo por su rapidez, escalabilidad y capacidad de formar grupos esféricos consistentes. 

Para definir el número ideal de segmentos, se validó el modelo K-Means mediante el **Método del Codo** y el **Coeficiente de Silueta**, los cuales confirmaron matemáticamente que **4 clústeres** era la división óptima.

## 4. Perfiles de Clientes Identificados
El modelo estructuró a la base de clientes en cuatro segmentos bien diferenciados:

* **Segmento 1: Los VIPs (Alto Valor)**
    * *Perfil:* Ingresos altos, ticket promedio elevado y visitas frecuentes.
    * *Comportamiento:* Compran en múltiples categorías, prefieren artículos caros y no dependen tanto de las promociones.
* **Segmento 2: Digitales y Tecnológicos**
    * *Perfil:* Gasto medio-alto, con fuerte inclinación a canales virtuales.
    * *Comportamiento:* Mayor tiempo de uso en la app y alta tasa de clics en campañas digitales. Su gasto se concentra en Electrónica.
* **Segmento 3: Ahorradores Recurrentes**
    * *Perfil:* Visitan la tienda física con mucha frecuencia, pero mantienen un ticket de compra bajo o moderado.
    * *Comportamiento:* Compras enfocadas en artículos de primera necesidad (Alimentos y Hogar). Buscan eficiencia y buen precio.
* **Segmento 4: Ocasionales / Riesgo de Fuga**
    * *Perfil:* Baja frecuencia de visitas y gasto mínimo.
    * *Comportamiento:* Utilizan la tienda de forma esporádica o para emergencias. Tienen la menor lealtad a la marca.

## 5. Recomendaciones Estratégicas
Se recomiendan las siguientes acciones directas por departamento:

**Para Marketing Comercial:**
* Suspender el envío del mismo boletín de ofertas masivo a todos los clientes.
* Enfocar el presupuesto publicitario digital y notificaciones de la app exclusivamente en el *Segmento 2*, destacando tecnología.
* Crear un programa de lealtad con beneficios exclusivos para el *Segmento 1*, asegurando su retención.

**Para Ventas y Operaciones:**
* Implementar un modelo de "compra programada" o suscripciones para alimentos básicos y limpieza, asegurando el retorno del *Segmento 3* y subiendo un poco su gasto.
* Diseñar campañas agresivas de descuento por tiempo limitado para el *Segmento 4* para incentivar una nueva visita.

**Para Logística e Infraestructura:**
* Ajustar el stock de las sucursales físicas analizando qué perfil predomina en la zona. Por ejemplo, reducir el espacio en góndola de tecnología y ampliar el de alimentos básicos en zonas dominadas por el *Segmento 3*.

# Justificación de Decisiones Técnicas

## 1. Preprocesamiento y Reducción Dimensional
El conjunto de datos original tenía 11 variables. Para evitar problemas de rendimiento y exceso de información cruzada, aplicamos el **Análisis de Componentes Principales (PCA)**. 

* **Decisión:** Nos quedamos con los componentes que explican al menos el 80% de la varianza. Esto nos ayudó a limpiar el ruido y hacer que los modelos funcionen de forma más eficiente.
* **Visualización:** Usamos **t-SNE** exclusivamente para proyectar los datos en gráficos 2D y que la gerencia pueda entender los grupos de forma visual. No utilizamos esta transformación como entrada para los algoritmos de clustering.

## 2. Selección del Algoritmo de Clustering
Pusimos a prueba tres modelos distintos:

* **Agrupamiento Jerárquico:** Lo descartamos como modelo principal porque consume demasiada memoria y tiempo. Ejecutarlo con la base completa representaba un riesgo de saturación, y limitaba mucho el proyecto si a futuro la empresa necesita analizar bases más grandes.
* **DBSCAN:** Es muy útil para detectar formas irregulares, pero su mecánica de aislar clientes atípicos (clasificándolos como ruido) terminó fragmentando grupos que comercialmente tenían sentido y afectó el puntaje de cohesión general de los grupos al evaluarlos con el Coeficiente de Silueta.
* **Decisión Final (K-Means):** Elegimos K-Means como el modelo maestro porque es rápido, escala sin problemas y forma grupos esféricos bien definidos. Desde el punto de vista del negocio, estos perfiles son mucho más fáciles de interpretar y transformar en campañas de marketing.

## 3. Determinación del Número de Clústeres (k)
Para no elegir la cantidad de grupos al azar, respaldamos la decisión con dos métodos matemáticos:

* **Método del Codo (Elbow Method):** Mostró un quiebre y estabilización claros al llegar a 4 clústeres.
* **Coeficiente de Silueta:** Alcanzó su puntaje máximo de cohesión y separación en $k=4$, validando la elección.
* **Sentido Comercial:** Dividir a los clientes en 4 grupos es el equilibrio ideal. Nos da el detalle necesario para armar campañas personalizadas sin fragmentar tanto la base de datos como para que la operación diaria del equipo se vuelva inmanejable.
