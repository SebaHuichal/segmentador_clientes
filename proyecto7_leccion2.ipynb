# Proyecto: Segmentador Inteligente de Clientes Minoristas
## Lección 2: Técnicas de Clusterización

### A. Definir escenarios donde se aplique clusterización

En el contexto de **Retail Insights S.A.**, la clusterización no busca predecir un número, sino descubrir la estructura natural de los datos. Estos son los escenarios clave donde aplicaremos esta técnica:

1. **Segmentación de Clientes (Perfilado Conductual):** Es el uso principal. Agrupamos a los clientes según su frecuencia de compra, gasto promedio y categorías preferidas. Esto permite pasar de un marketing masivo a uno personalizado.
   

2. **Agrupamiento de Productos (Product Bundling):** Identificar qué productos tienen un comportamiento de venta similar o son comprados por el mismo tipo de cliente. Sirve para diseñar "combos" o mejorar el acomodo en las góndolas de las tiendas físicas.
   

3. **Análisis de Ubicaciones (Tiendas):** Clasificar las sucursales de la cadena en grupos según su rendimiento, tráfico de gente y horarios pico. Ayuda a la consultora a recomendar estrategias de logística diferenciadas por zona.


4. **Identificación de Clientes en Riesgo (Churn Prevention):** Detectar grupos de clientes que muestran una disminución inusual en su actividad. Al agruparlos, podemos entender qué les molesta y actuar antes de que dejen de comprar.


5. **Optimización de Precios:** Agrupar productos que son sensibles al precio (elásticos) frente a los que no lo son, permitiendo ajustar márgenes sin perder volumen de ventas.


### B. Ventajas y desventajas de la técnica

Para tomar decisiones informadas en la consultoría **Retail Insights S.A.**, es vital entender los puntos fuertes y las limitaciones de los algoritmos de agrupamiento.

| Aspecto | Ventajas  | Desventajas  |
| :--- | :--- | :--- |
| **Descubrimiento** | Encuentra segmentos y patrones ocultos que las reglas de negocio manuales no logran ver. | La interpretación es subjetiva; el algoritmo agrupa puntos, pero el humano debe darles sentido de negocio. |
| **Etiquetado** | No requiere datos etiquetados previamente, lo que ahorra tiempo y costos de supervisión. | No hay una "respuesta correcta" (ground truth), lo que dificulta medir la precisión exacta del modelo. |
| **Flexibilidad** | Se adapta a múltiples variables (edad, gasto, visitas) y a la evolución de los datos en el tiempo. | Es extremadamente sensible a la escala de los datos; si no normalizamos bien, una variable opacará a las demás. |
| **Parámetros** | Algoritmos como DBSCAN detectan automáticamente el ruido y valores atípicos (outliers). | Técnicas como K-Means exigen que el usuario decida el número de grupos ($k$) antes de empezar. |
| **Acción Comercial**| Permite crear estrategias de marketing personalizadas para miles de clientes a la vez. | Algunos métodos son costosos computacionalmente y se vuelven lentos si la base de clientes crece masivamente. |

### C. Comparar K-Means, DBSCAN y agrupamiento jerárquico teóricamente

Para elegir el algoritmo correcto, debemos entender que cada uno "mira" a los clientes de una forma distinta. Aquí tienes la tabla comparativa:

| Criterio | K-Means | DBSCAN | Agrupamiento Jerárquico |
| :--- | :--- | :--- | :--- |
| **Cómo funciona** | Divide los datos en "k" grupos buscando el centro (centroide) de cada uno. | Agrupa puntos que están muy cerca (densidad) y separa los que están aislados. | Crea un árbol de similitudes (dendrograma) conectando puntos de abajo hacia arriba. |
| **Forma de los grupos** | Solo encuentra grupos circulares o esféricos. | Puede encontrar grupos de cualquier forma (alargados, en arco, etc.). | Muy flexible, depende de cómo se decida cortar el árbol. |
| **Parámetros clave** | Requiere que tú le digas el número de grupos ($k$). | Requiere la distancia máxima ($eps$) y el mínimo de puntos ($min\_samples$). | Requiere elegir la métrica de distancia y el método de enlace. |
| **Manejo de ruido** | Muy sensible a outliers (un cliente que gasta demasiado moverá todo el centroide). | Excelente; identifica y marca los valores atípicos como "ruido". | Sensible; un valor extraño puede alterar toda la jerarquía de conexiones. |
| **Escalabilidad** | Muy rápido; ideal para bases de datos muy grandes. | Lento con muchos datos, pero eficiente en detectar formas complejas. | Muy lento; no se recomienda para datasets con miles de registros. |

> ⚠️ **Consideración práctica para este proyecto:** El agrupamiento jerárquico es computacionalmente costoso y no se recomienda para datasets con miles de registros. En este proyecto, el dataset `clientes_alta_dimension.csv` contiene 2000 registros. Para poder aplicar este algoritmo de forma obligatoria (según los requerimientos del módulo) sin comprometer el rendimiento, **se utilizará una muestra aleatoria estratificada de 300 a 500 registros** para el análisis jerárquico. Los resultados de K-Means y DBSCAN sí se aplicarán sobre el dataset completo. Esta decisión quedará documentada y justificada en el informe final.


### D. Asociar cada algoritmo a un tipo de dato o problema real

No todos los algoritmos sirven para lo mismo. En **Retail Insights S.A.**, elegiremos la herramienta según el tipo de datos que tengamos y lo que queramos descubrir:

| Algoritmo | Tipo de Datos Ideal | Ejemplo de Problema Real |
| :--- | :--- | :--- |
| **K-Means** | Datos numéricos bien definidos y con varianza similar (ej. Edad vs. Gasto Total). | **Niveles de Lealtad:** Clasificar a los clientes en "Bronce, Plata y Oro" basándose en su frecuencia de compra y monto gastado. |
| **DBSCAN** | Datos con mucho ruido, valores atípicos o formas geográficas/irregulares. | **Detección de Fraude:** Identificar transacciones sospechosas que se alejan del comportamiento normal o agrupar tiendas físicas por su ubicación GPS real. |
| **Agrupamiento Jerárquico** | Datasets pequeños donde importa entender la relación de "padre e hijo" entre grupos. | **Taxonomía de Productos:** Crear categorías y subcategorías de productos (ej. Hogar -> Cocina -> Utensilios) basándose en qué tan seguido se compran juntos. |



#### ¿Cuándo usar cada uno en nuestro proyecto?
* Usaremos **K-Means** como nuestro motor principal para la segmentación básica de clientes porque es rápido y fácil de explicar a los gerentes.
* Recurriremos a **DBSCAN** si notamos que los "outliers" (clientes con comportamientos muy raros) están ensuciando mucho nuestros grupos.
* El **Jerárquico** lo dejaremos para cuando necesitemos explicarle a la consultora cómo un segmento grande de "Jóvenes" se subdivide en subgrupos más específicos como "Gamer" o "Deportista".

> 📌 **Nota de implementación — Agrupamiento Jerárquico:** Dado que este algoritmo tiene complejidad O(n²) en memoria y tiempo, al aplicarlo sobre el dataset completo (2000 registros) puede ser muy lento o generar errores de memoria. **La estrategia recomendada es submuestrear a ~400 registros** antes de ejecutarlo, dejando constancia de esto en el código con un comentario explicativo. K-Means y DBSCAN no tienen esta restricción y se ejecutan sobre el dataset completo.
