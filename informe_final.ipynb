# Informe Final: Segmentación Inteligente de Clientes
**Cliente:** Retail Insights S.A.
**Departamento:** Inteligencia de Negocios

## 1. Introducción y Objetivo del Proyecto
Retail Insights S.A. identificó la necesidad de abandonar las estrategias de marketing genérico masivo para uno de sus clientes principales, una cadena con presencia en tiendas físicas y e-commerce. Hasta la fecha, las campañas se basaban en información básica, perdiendo la oportunidad de capitalizar los hábitos reales de consumo.

El objetivo de este proyecto fue desarrollar un modelo de segmentación basado en algoritmos de aprendizaje no supervisado. Al analizar una base de 2000 clientes con 11 variables de comportamiento, demográficas y de interacción digital, el sistema logró descubrir patrones ocultos para crear perfiles accionables, permitiendo la personalización de futuras estrategias comerciales.

## 2. Metodología y Tratamiento de Datos
Para garantizar que el análisis fuera preciso y escalable, se aplicó un flujo de trabajo riguroso:

* **Limpieza y Preparación:** Se estructuró un conjunto de datos multidimensional que incluía ingresos, frecuencia de visitas, gasto por categoría (electrónica, hogar, ropa, etc.) y métricas de uso de la aplicación móvil.
* **Reducción de Dimensionalidad (PCA):** Trabajar con 11 variables simultáneas genera "ruido" computacional. Se aplicó el Análisis de Componentes Principales (PCA), conservando únicamente los componentes necesarios para explicar más del 80% de la varianza de los datos. Esto optimizó el rendimiento de los modelos.
* **Visualización (t-SNE):** Para que los resultados fuesen interpretables visualmente por la gerencia, se utilizó la técnica t-SNE, proyectando la complejidad multidimensional en gráficos 2D claros y definidos.

## 3. Selección y Validación del Modelo Maestro
Se pusieron a prueba tres enfoques de clusterización para entender cuál representaba mejor la realidad del negocio:

1.  **Agrupamiento Jerárquico:** Descartado para el modelo final por su alto consumo de memoria computacional, lo que limitaba su escalabilidad futura si la base de clientes crecía.
2.  **DBSCAN:** Descartado porque su mecánica de aislar clientes atípicos como ruido terminó fragmentando grupos que comercialmente tenían sentido y afectó negativamente el puntaje de cohesión general al evaluarlos con el Coeficiente de Silueta.
3.  **K-Means (Modelo Seleccionado):** Elegido como el algoritmo definitivo por su rapidez, escalabilidad y capacidad de formar grupos esféricos consistentes. 

Para definir el número ideal de segmentos sin sesgos arbitrarios, se validó el modelo K-Means mediante el **Método del Codo** y el **Coeficiente de Silueta**, los cuales confirmaron matemáticamente que **4 clústeres** era la división óptima para la base de datos.

## 4. Perfiles de Clientes Identificados
El modelo estructuró a la base de clientes en cuatro segmentos bien diferenciados. A continuación se detalla su comportamiento promedio:

* **Segmento 1: Los VIPs (Alto Valor)**
    * *Perfil:* Clientes con ingresos altos, ticket promedio elevado y visitas frecuentes.
    * *Comportamiento:* Compran en múltiples categorías, con preferencia por artículos de ticket alto. No son altamente dependientes de las promociones.
* **Segmento 2: Digitales y Tecnológicos**
    * *Perfil:* Nivel de gasto medio-alto, pero con una fuerte inclinación a los canales virtuales.
    * *Comportamiento:* Registran el mayor tiempo de uso en la aplicación móvil y alta tasa de clics en campañas digitales. Su gasto se concentra fuertemente en la categoría de Electrónica.
* **Segmento 3: Ahorradores Recurrentes**
    * *Perfil:* Visitan la tienda física con mucha frecuencia, pero mantienen un ticket de compra bajo o moderado en cada visita.
    * *Comportamiento:* Sus compras están enfocadas casi exclusivamente en artículos de primera necesidad, como Alimentos y productos para el Hogar. Buscan eficiencia y precio.
* **Segmento 4: Ocasionales / Riesgo de Fuga**
    * *Perfil:* Baja frecuencia de visitas y gasto mínimo.
    * *Comportamiento:* Utilizan la tienda de forma esporádica o para emergencias. Es el segmento con menor lealtad a la marca.

## 5. Recomendaciones Estratégicas
Gracias a esta segmentación, se recomiendan las siguientes acciones directas por departamento para maximizar la rentabilidad:

**Para Marketing Comercial:**
* **Fin de los envíos masivos:** Suspender el envío del mismo boletín de ofertas a los 2000 clientes.
* **Campaña Digitales:** Enfocar el presupuesto publicitario digital y las notificaciones *push* de la app exclusivamente en el *Segmento 2*, destacando lanzamientos de tecnología.
* **Fidelización VIP:** Crear un programa de lealtad ("Status Elite") con beneficios exclusivos y acceso anticipado a productos para el *Segmento 1*, asegurando su retención.

**Para Ventas y Operaciones:**
* **Suscripciones para Ahorradores:** Implementar un modelo de "compra programada" o cajas de suscripción mensual para alimentos básicos y limpieza, asegurando el retorno del *Segmento 3* y subiendo ligeramente su gasto mensual.
* **Reactivación de Ocasionales:** Diseñar campañas agresivas de descuento por tiempo limitado dirigidas al *Segmento 4* para incentivar una segunda o tercera visita, midiendo su tasa de conversión antes de descartarlos.

**Para Logística e Infraestructura:**
* **Inventario Inteligente:** Ajustar el stock de las sucursales físicas analizando qué perfil predomina en cada código postal. Por ejemplo, reducir el espacio en góndola de tecnología y ampliar el de alimentos básicos en zonas dominadas por el *Segmento 3*.
